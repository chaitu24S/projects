```python
import requests
import psycopg2

# PostgreSQL Connection
conn = psycopg2.connect(
    host="localhost",
    database="learning_db",
    user="postgres",
    password="admin"
)

cursor = conn.cursor()
```

```python
print(conn.get_dsn_parameters())
```

```python
# Extract

url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)
users = response.json()
```

```python
# Transform & Load

for user in users:
    user_id = user["id"]
    name = user["name"]
    email = user["email"]
    company = user["company"]["name"]

    cursor.execute("""
    INSERT INTO users (id, name, email, company)
    VALUES (%s, %s, %s, %s)              
""", (user_id, name, email, company))
```

```python
# Save changes
conn.commit()
```

```python
# Close connections
cursor.close()
conn.close()

print("Data loaded successfully")
```
